# Data Analysis & Visualization of Diamond Dataset

Source: https://www.kaggle.com/datasets/nancyalaswad90/diamonds-prices/data

```python
# Importing the necessary packages 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data into a datframe
df = pd.read_csv('Diamonds_Prices2022.csv')
```

```python
df.head()
```

The dataset contains an unnamed column so we drop that column.

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
# Printing the first five rows
df.head()
```

Looking at the content of datset we see that columns x, y and z are as follows:
    
x length in mm (0--10.74)

y width in mm (0--58.9)

z depth in mm (0--31.8)

```python
max_price = df.loc[df['price'].argmax()]

# Print features of max price diamond
print('Max Price: ', max_price['price'])
print('Carat: ', max_price['carat'])
print('Cut: ', max_price['cut']) 
print('Color: ', max_price['color'])
print('Clarity: ', max_price['clarity'])
print('Depth: ', max_price['depth'])  
print('Table: ', max_price['table'])
print('X: ', max_price['x'])
print('Y: ', max_price['y'])
print('Z: ', max_price['z'])
```

```python
min_price = df.loc[df['price'].argmin()]  

# Print features of min price diamond
print('Min Price: ', min_price['price'])
print('Carat: ', min_price['carat'])
print('Cut: ', min_price['cut'])
print('Color: ', min_price['color'])  
print('Clarity: ', min_price['clarity'])
print('Depth: ', min_price['depth'])
print('Table: ', min_price['table'])
print('X: ', min_price['x'])
print('Y: ', min_price['y'])
print('Z: ', min_price['z'])
```

```python
sns.set()
```

```python
cut_list = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.catplot(data=df, x='cut', kind='count', order=cut_list)
plt.title('Diamonds by cut')
plt.show()
```

```python
df['color'].unique()
```

```python
# Groupby cut and color
df.groupby(['cut', 'color']).size()
```

```python
grouped = df.groupby(['cut', 'color'])

# Calculating the average price for each group
avg_price = grouped['price'].mean()

print(avg_price)
```

```python
sns.countplot(data=df, x='cut', hue='color')
plt.title("# of Diamonds by Cut and Color")
plt.show()
```

```python
sns.catplot(data=df, x='clarity', y='price', kind='point')
plt.title("Diamond price by clarity")
plt.show()
```

```python
sns.catplot(data=df, x='cut', y='price', kind='point', col='clarity')
plt.tight_layout()
plt.show()
```

```python
sns.catplot(data=df, x='price', y='cut', kind='box', sym='')
plt.show()
```

```python
plt.figure(figsize=(14, 28))
sns.violinplot(data=df, y='cut', x='price', hue='color')
plt.show()
```

```python
s = sns.FacetGrid(df, col='cut')
s.map(sns.boxplot, 'price', order=cut_list)
plt.show()
```

```python
sns.lmplot(data=df, x='carat', y='price', hue='cut', col='cut')
plt.show()
```

```python
# Finding the correlation between the variables
df.corr()
```

```python
# Plotting the correlation heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
